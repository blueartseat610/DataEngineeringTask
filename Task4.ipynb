```python
# Start Spark
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("ETL Movie Ratings") \
    .getOrCreate()
```

```python
users_df = spark.read.csv("users.csv", header=True, inferSchema=True)
movies_df = spark.read.csv("movies.csv", header=True, inferSchema=True)

users_df.show()
movies_df.show()
```

```python
import pandas as pd
import numpy as np

ratings = pd.read_csv("ratings.csv")

# get unique users
user_ids = ratings["userId"].unique()

# create fake demographic data
np.random.seed(42)

ages = np.random.randint(18,60,len(user_ids))
genders = np.random.choice(["M","F"],len(user_ids))
occupations = np.random.choice(
    ["Student","Engineer","Teacher","Manager","Doctor","Designer"],
    len(user_ids)
)

users = pd.DataFrame({
    "user_id": user_ids,
    "age": ages,
    "gender": genders,
    "occupation": occupations
})

users.to_csv("users.csv",index=False)

users.head()
```

```python
ratings = pd.read_csv("ratings.csv")

ratings = ratings.rename(columns={
    "userId":"user_id",
    "movieId":"movie_id"
})

ratings.to_csv("ratings_clean.csv",index=False)
```

```python
users_df = spark.read.csv("users.csv",header=True,inferSchema=True)
movies_df = spark.read.csv("movies.csv",header=True,inferSchema=True)
ratings_df = spark.read.csv("ratings_clean.csv",header=True,inferSchema=True)

users_df.show(5)
movies_df.show(5)
ratings_df.show(5)
```
